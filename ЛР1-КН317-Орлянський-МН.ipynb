```python
from itertools import product
from scipy import signal

def generate_mines(field_width, field_height, num_mines):
  all_positions = list(product(range(field_height), range(field_width)))
  np.random.shuffle(all_positions)
  all_positions = np.array(all_positions[:num_mines]) 
  field = np.zeros(shape=(field_height, field_width), dtype=int)
  field[all_positions.T[0], all_positions.T[1]] = 1
  return field

def mask_to_field(field_mask):
  ker = np.ones(shape=(3, 3), dtype=np.uint8)
  neighbours = signal.convolve2d(field_mask, ker, mode='same', boundary='fill', fillvalue=0)
  neighbours[field_mask == 1] = -1
  return neighbours

field_width = 5 
field_height = 3 
num_mines = 3 
field_mask = generate_mines(field_width, field_height, num_mines)
field = mask_to_field(field_mask)
field
```

Тепер оскільки z-score рахується як:

Де:
x - значення ознак з набору даних
μ - середне значення для даної онзнаки x
σ - стандартне відхилення

То аби порахувати Z-score створимо клас, який буде приймати значення ознах, та буде повертати значення Z-score для даних ознак

```python
class Zscore:
   def __init__(self, np_columns: np.ndarray):
       not_nan_columns = np_columns[~np.isnan(np_columns)]
       self.mean = np.mean(not_nan_columns)
       self.std = np.std(not_nan_columns)
       self.x = np_columns
    
   def get_score(self):
       return np.where(np.isnan(self.x), np.nan, np.abs((self.x - self.mean) / self.std))
```

Аби мати куди записувати результати Z-score і не переписувати початковий датафрейм створимо новий пустий датаферйм, де будуть зберігатися всі результати Z-score для кожної з ознак:

```python
penguins_z_score_df = pd.DataFrame()
```

На основі створеного класу дослідимо Z-score для заданих ознак та запишимо їхні значення в penguins_z_score_df:

```python
for column in penguins.select_dtypes(include=[np.number]).columns:
    z_score = Zscore(penguins[column].to_numpy())
    penguins_z_score_df[column] = z_score.get_score()
```
